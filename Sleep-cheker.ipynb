```python
import uuid
import os
import time
```

```python
ima_path= os.path.join('data','images')
labesl=['awake','Sleepy']
number_im=20
```

```python
cap=cv2.VideoCapture(0)
for label in labesl:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    for im_no in range(number_im):
        print('Collecting images for {}, images number {}'.format(label,im_no))
        ret,frame=cap.read()
        imname=os.path.join(ima_path,label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(imname,frame)
```
